```python
import pandas as pd # to import the data from various file source
from sklearn.preprocessing import LabelEncoder   # to convert non numeric into numeric values
from sklearn.model_selection import train_test_split   # to split the data into train and test purpose
from sklearn.ensemble import RandomForestClassifier     # model
from sklearn.metrics import accuracy_score,fbeta_score,precision_score,recall_score   # accuracy check
```

```python
data = pd.read_csv(r'/content/dataset.csv')
data
```

```python
data.info()       # to check the data type
```

```python
lbl_ed = LabelEncoder()
```

```python
data['work'] = lbl_ed.fit_transform(data['work'])    # to convert object data into int for work colum like that do for all
```

```python
data['goal'] = lbl_ed.fit_transform(data['goal'])
```

```python
data['learning'] = lbl_ed.fit_transform(data['learning'])
```

```python
data['output'] = lbl_ed.fit_transform(data['output'])
```

```python
data.info()
```

```python
x = data.drop(columns=['output'])
y = data['output']
```

```python
x
```

```python
y
```

Splitting the data as train and test

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=10)
```

model

```python
random = RandomForestClassifier(n_estimators=100,random_state=10,max_depth=5)
```

```python
random.fit(x_train,y_train)
```

```python
random_pred = random.predict(x_test)
```

```python
random_pred
```

```python
fbeta_score(y_test,random_pred,average='micro',beta=1)
```
